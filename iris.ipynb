```python
fig = iris[iris.Species=='Iris-setosa'].plot.scatter(x='SepalLengthCm', y= 'SepalWidthCm', color='orange', label='Setosa')
iris[iris.Species=='Iris-versicolor'].plot.scatter(x='SepalLengthCm', y= 'SepalWidthCm', color='blue', label='Versicolor',ax=fig)
iris[iris.Species=='Iris-virginica'].plot.scatter(x='SepalLengthCm', y= 'SepalWidthCm', color='red', label='Virginica',ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal width")
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

```python
fig = iris[iris.Species=='Iris-setosa'].plot.scatter(x='PetalLengthCm', y='PetalWidthCm', color='orange', label='Setosa')
iris[iris.Species=='Iris-versicolor'].plot.scatter(x='PetalLengthCm', y='PetalWidthCm', color='red', label='Versicolor', ax=fig)
iris[iris.Species=='Iris-virginica'].plot.scatter(x='PetalLengthCm', y='PetalWidthCm', color='green', label='Virginica', ax=fig)
fig.set_xlabel('Petal Length')
fig.set_ylabel('Petal Width')
fig = plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

```python
from sklearn.cross_validation import train_test_split
from sklearn import svm
from sklearn import metrics
```

```python
train, test = train_test_split(iris, test_size=.3)
```

```python
print(train.shape)
test.shape
```

```python
train_x = train[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
train_y = train.Species
test_x = train[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
test_y = train.Species
```

```python
train_x.head(2)
```

```python
train_y.head(2)
```

```python
model = svm.SVC()
model.fit(train_x, train_y)
prediction = model.predict(test_x)
print('Accuracy of SVM :', metrics.accuracy_score(prediction, test_y))
```
